### QUESTION 2:

```python
# Reading all state files and converting them into list of dataframes
# Concatenating all dataframes into one dataframe
import pandas as pd
import numpy as np

import glob
filenames=glob.glob('state*.csv')
dataframes=[]
for filename in filenames:
    df=pd.read_csv(filename,index_col=0,header=0)
    dataframes.append(df)
us_census=pd.concat(dataframes,axis=0,ignore_index=True) 
```

### QUESTION 3:

```python
# Specifying datatypes and column names of us_census dataframe
print("DATATYPES:","\n\n",us_census.dtypes,"\n")
print("COLUMNS:","\n\n",us_census.columns)
```

<b>Are those datatypes going to hinder you as you try to make histograms?</b>

Yes, data need to be either float or integer for making histograms

### QUESTION 4:

```python
# Columns like 'Hispanic','White','Black','Native','Asian','Pacific' are object because of the presence of non-numeric '%' sign.
# Column 'Income' is non-numeric because of '$' sign.
# Column GenderPop has non-numeric 'M','F' and '_' sign.
# The above mentioned non-numeric signs need to be removed for converting data to either integer or float.
us_census.head()
```

### QUESTION 5:

```python
us_census['Income'].replace('\$','',regex=True,inplace=True)
```

### QUESTION 6:

```python
# Split GenderPop column into Male and Female column
us_census[['Male','Female']]=us_census['GenderPop'].str.split('_',n=-1,expand=True)
```

```python
# Newly created male and female column are object datatype
us_census.dtypes
```

### QUESTION 7:

```python
# Remove 'M' from Male column and 'F' from Female column
# Both Male and Female column are object types
us_census['Male'].replace('M','',regex=True,inplace=True)
us_census['Female'].replace('F','',regex=True,inplace=True)
us_census['Female'].replace('',np.nan,regex=True,inplace=True)
```

### QUESTION 8:

```python
# Average income in a state vs proportion of women in that state
# Both Female and Income column are object types
# ScatterPlot gives error because of the presence of nan values in Female Column
from matplotlib import pyplot as plt
plt.scatter(us_census['Female'],us_census['Income'])
plt.show()
```

### QUESTION 9:

```python
# Female column has nan values
print(us_census['Female'])

# Convert Male column to int type
us_census['Male']=us_census['Male'].astype('int64')

# Fill nan values in Female column, then convert to int type
us_census['Female'].fillna(us_census['TotalPop']-us_census['Male'],inplace=True)
us_census['Female']=us_census['Female'].astype('int64')

# Both 'Male' and 'Female' column are converted to int datatype
print(us_census.dtypes)

# Nan values in Female column have been filled with (TotalPop-Male) population
print(us_census['Female'])
```

### QUESTION 10

### QUESTION 11

```python
# Drop duplicates in us_census dataframe
us_census.drop_duplicates(inplace=True)
```

### QUESTION 12:

```python
# Average income in a state vs proportion of women in that state
# Female column is int type and Income column is converted below to float type
from matplotlib import pyplot as plt
us_census['Income']=us_census['Income'].astype('float64')
plt.scatter(us_census['Female'],us_census['Income'])
plt.show()
```

### QUESTON 13

```python
us_census.columns
```

Race columns are 'Hispanic', 'White', 'Black', 'Native', 'Asian' and 'Pacific'

### QUESTION 14:

```python
# Replacing '%' sign in race columns
us_census['Hispanic'].replace('\%','',regex=True,inplace=True)
us_census['White'].replace('\%','',regex=True,inplace=True)
us_census['Black'].replace('\%','',regex=True,inplace=True)
us_census['Native'].replace('\%','',regex=True,inplace=True)
us_census['Asian'].replace('\%','',regex=True,inplace=True)
us_census['Pacific'].replace('\%','',regex=True,inplace=True)
```

```python
# Convert race columns to float64 datatype
us_census=us_census.astype({'Hispanic':'float64','White':'float64','Black':'float64','Native':'float64','Asian':'float64','Pacific':'float64'})
```

```python
# Fill nan values with mean in race column
us_census=us_census.apply(lambda x: x.fillna(x.mean()) if x.name in ['Hispanic', 'White', 'Black', 'Native', 'Asian','Pacific'] else x)
```

```python
# See if any duplicated rows are left
us_census.duplicated()
# No more duplicated rows now
```

```python
# histogram of race columns
us_census.hist(column=['Hispanic', 'White', 'Black', 'Native', 'Asian','Pacific'])
```
